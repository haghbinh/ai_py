<h2 dir="rtl" style="text-align: justify;">
    متغیر چیست؟
</h2>
<p dir="rtl" style="text-align: justify;">
متغیر (Variable) محفظه‌ای است برای نگهداری داده‌ها. در پایتون، نیازی به تعیین نوع متغیر نیست و نوع آن به صورت خودکار با توجه به مقدار تعیین می‌شود.
</p>

```python
x = 5        # عدد صحیح (int)
pi = 3.14    # عدد اعشاری (float)
name = 'Ali' # رشته (string)
flag = True  # مقدار بولی (bool)

print(x, pi, name, flag)
```

<h2 dir="rtl" style="text-align: justify;">
 بررسی نوع داده
</h2>
<p dir="rtl" style="text-align: justify;">
برای بررسی نوع یک متغیر می‌توان از تابع `type` استفاده کرد:
</p>

```python
print(type(x))
print(type(pi))
print(type(name))
print(type(flag))
```

<h2 dir="rtl" style="text-align: justify;">
انواع داده عددی در پایتون
</h2>

```python
a = 10       # int
b = 3.5      # float

print(type(a), type(b))
```

<h2 dir="rtl" style="text-align: justify;">
رشته‌ها (Strings)
</h2>
<p dir="rtl" style="text-align: justify;">
رشته‌ها مجموعه‌ای از کاراکترها هستند که در داخل کوتیشن ('' یا "") قرار می‌گیرند.
</p>

```python
text = "Hello"
print(text)
print(text.upper())
print(text.lower())
print(text[0])
```

<h2 dir="rtl" style="text-align: justify;">
مقادیر بولی (Boolean)
</h2>
<p dir="rtl" style="text-align: justify;">
نوع `bool` دو مقدار دارد: `True` و `False`. این نوع در تصمیم‌گیری‌ها و شرط‌ها استفاده می‌شود.
</p>

```python
a = 10
b = 5
print(a > b)    # True
print(a == b)   # False
```

<h2 dir="rtl" style="text-align: justify;">۱. تابع <code>input()</code></h2>

<p dir="rtl" style="text-align: justify;">
تابع <b>input()</b> برای دریافت داده از کاربر استفاده می‌شود. هر چیزی که کاربر وارد کند به صورت <b>رشته (string)</b> دریافت می‌شود.
</p>

```python
name = input("لطفاً نام خود را وارد کنید: ")
print("سلام،", name)
```

<p dir="rtl" style="text-align: justify;"> اگر بخواهیم عدد دریافت کنیم، باید رشته را به نوع عددی تبدیل کنیم: </p>

```python
age = int(input("سن خود را وارد کنید: "))
print("شما", age, "سال دارید.")
```

<h2 dir="rtl" style="text-align: justify;">۲. تابع <code>print()</code></h2> <p dir="rtl" style="text-align: justify;"> تابع <b>print()</b> برای نمایش خروجی در کنسول استفاده می‌شود. می‌توان چند مقدار را با کاما جدا کرد یا متن و متغیر را با هم چاپ کرد. </p>

```python
a = 5
b = 10
print("جمع a و b برابر است با:", a + b)
```

<h2 dir="rtl" style="text-align: justify;">۳. f-string</h2> <p dir="rtl" style="text-align: justify;"> از نسخه‌ی پایتون ۳.۶ به بعد، می‌توانیم از <b>f-string</b> برای ترکیب متن و متغیرها به صورت ساده و خوانا استفاده کنیم. </p>

```python
name = "علی"
age = 25
print(f"سلام {name}! شما {age} ساله هستید.")
```

<h2 dir="rtl" style="text-align: justify;">کنترل جریان (Control Flow)</h2>

<p dir="rtl" style="text-align: justify;">
در برنامه‌نویسی، <b>کنترل جریان</b> به تصمیم‌گیری و تکرار در اجرای دستورات گفته می‌شود. با استفاده از شرط‌ها، حلقه‌ها و دستورات <b>break</b> و <b>continue</b> می‌توان مسیر اجرای برنامه را کنترل کرد.
</p>

<h2 dir="rtl" style="text-align: justify;">۱. شرط‌ها (if, elif, else)</h2>

<p dir="rtl" style="text-align: justify;">
با دستور <b>if</b> می‌توانیم بر اساس یک شرط تصمیم بگیریم که بخش خاصی از کد اجرا شود. از <b>elif</b> و <b>else</b> برای بررسی شرایط دیگر یا حالت پیش‌فرض استفاده می‌کنیم.
</p>

```python
x = 15

if x > 20:
    print("x is greater than 20")
elif x == 15:
    print("x is exactly 15")
else:
    print("x is less than 15")
```

<h2 dir="rtl" style="text-align: justify;">۲. حلقه‌ها (Loops)</h2> <p dir="rtl" style="text-align: justify;"> در پایتون دو نوع حلقه داریم: <b>for</b> و <b>while</b>. - حلقه <b>for</b> برای تکرار روی یک مجموعه (لیست، رشته، یا رنج عددی) استفاده می‌شود. - حلقه <b>while</b> تا زمانی که شرط برقرار است، اجرا می‌شود. </p>

```python
# for loop example
for i in range(5):
    print("Iteration:", i)
```

```python

# while loop example
i = 0
while i < 5:
    print("Iteration:", i)
    i += 1
```

### Nested Loops

```python
for i in [1, 2]:
    for j in [10, 20, 30]:
        print(i, j)
```

<h2 dir="rtl" style="text-align: justify;">۳. دستور <code>break</code></h2> <p dir="rtl" style="text-align: justify;"> با دستور <b>break</b> می‌توان اجرای حلقه را قبل از اتمام کامل آن متوقف کرد. </p>

```python
for i in range(10):
    if i == 5:
        break
    print("Value:", i)
```

<h2 dir="rtl" style="text-align: justify;">۴. دستور <code>continue</code></h2> <p dir="rtl" style="text-align: justify;"> با دستور <b>continue</b> می‌توان اجرای حلقه را در همان تکرار متوقف کرد و به تکرار بعدی رفت. </p>

```python
for i in range(5):
    if i == 2:
        continue
    print("Number:", i)
```

<h2 dir="rtl" style="text-align: justify;">مثال:</h2>
 <p dir="rtl" style="text-align: justify;">
برنامه ای بنویسید که مثلث های زیر را چاپ کند:
</p>

*</br>
** </br>
*** </br>
**** </br>
***** </br>
